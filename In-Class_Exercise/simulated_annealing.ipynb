```python
import numpy as np
import matplotlib.pyplot as plt

# initial temperature
T = 10.0
x = np.random.uniform(0, 100)

def x_function(x):
    return 3 * x**2 - 60 * x + 9

stop = 1e-7  # stop criterion

# histories for plotting
x_hist = []
score_hist = []
T_hist = []

i = 0
while T > stop:
    y = x_function(x)
    x_new = x + np.random.uniform(-1, 1)

    if 0 <= x_new <= 100:
        y_new = x_function(x_new)
        if y_new < y:  # find minimum value
            x = x_new
        else:
            p = np.exp((y - y_new) / T)
            r = np.random.uniform(0, 1)
            if p > r:
                x = x_new

        x_hist.append(x)
        T_hist.append(T)

    i += 1
    T *= 0.999
    print(x, x_function(x))

# --- Plot 1: x_function(x) for x in [0, 100]
xs = np.linspace(0, 100, 1000)
ys = x_function(xs)

# --- Plot 2 & 3: histories over iterations
iters = np.arange(len(x_hist))

fig, axes = plt.subplots(3, 1, figsize=(9, 10), constrained_layout=True)

axes[0].plot(xs, ys, color="black", lw=2)
axes[0].set_title("x_function(x) on [0, 100]")
axes[0].set_xlabel("x")
axes[0].set_ylabel("x_function(x)")
axes[0].grid(True, alpha=0.3)

axes[1].plot(iters, x_hist, lw=1.5)
axes[1].set_title("x over iterations")
axes[1].set_xlabel("iteration")
axes[1].set_ylabel("x")
axes[1].grid(True, alpha=0.3)

axes[2].plot(iters, T_hist, lw=1.5)
axes[2].set_title("Temperature T over iterations")
axes[2].set_xlabel("iteration")
axes[2].set_ylabel("T")
axes[2].grid(True, alpha=0.3)

plt.show()
```
